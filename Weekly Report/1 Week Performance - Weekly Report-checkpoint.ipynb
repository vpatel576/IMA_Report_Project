```python
import pandas as pd
import pandas_datareader as web
from datetime import datetime
from datetime import timedelta

import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
today = datetime.today().date()
```

```python
start = today - timedelta(days = 7)
```

```python
symbol = ['XLC', 'XLU', 'XLV', 'XLP', 'XLY', 'XLF', 'XLK', 'XLRE', 'XLB', 'XLE', 'XLI']
```

```python
industry = {}
```

```python
for i in symbol:
    close = web.DataReader(i, 'yahoo', start, today)['Adj Close']
    y = (((close[-1] - close[0])/(close[0]))*100)
    z = str(round(y,2))
    industry[i] = z
```

```python
industry
```

```python

x = dict(sorted(industry.items(), key=lambda item: item[1]))
v = x.values()
val = list(v)
abs_val = []
for i in val:
    abs_val.append(abs(float(i)))
abs_val
```

```python

plt.rcdefaults()
fig, ax = plt.subplots()

percent = val
tickers = x.keys()
y_pos = np.arange(len(tickers))



ax.barh(y_pos, percent, align='center', color='green', edgecolor = 'black')
ax.set_yticks(y_pos)
ax.set_yticklabels(tickers)
ax.set_xlabel('Percent Change (%)')
ax.set_title('1 Week Perfromance %')





plt.show()
```

```python

plt.rcdefaults()
fig, ax = plt.subplots()

percent = val
tickers = x.keys()
y_pos = np.arange(len(tickers))

clrs = []
for i in val:
    if float(i) < 0:
        clrs.append('red')
    else:
        clrs.append('green')

    
ax.barh(y_pos, percent, align='center', color=clrs, edgecolor = 'black')
ax.set_yticks(y_pos)
ax.set_yticklabels(tickers)
ax.set_xlabel('Percent Change (%)')
ax.set_ylabel('Tickers')
ax.set_title('1 Week Perfromance %')





plt.show()
```
